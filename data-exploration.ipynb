```python
import pandas as pd 
```

```python
data = pd.read_csv('books.csv')
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ax= plt.axes()
sns.heatmap(data.isna().transpose(),cbar=False, ax=ax)
plt.xlabel('Columns')
plt.ylabel('Missing Values')
plt.show()
```

```python
import numpy as np
data['missing-description'] = np.where(data['description'].isna(), 1, 0)
data['age-of-book'] = 2023 - data['published_year']
```

```python
columns_of_interset = ['num_pages', 'age-of-book', 'missing-description', 'average_rating']
correlation_matrix = data[columns_of_interset].corr(method='spearman')
sns.set_theme(style="white")
plt.Figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f',cbar_kws={'label': 'Spearman Correlation'})
heatmap.set_title('Correlation Matrix of Selected Features', fontdict={'fontsize':18}, pad=16)
plt.show()
```

```python
data[(data['description'].isna())|
     (data['num_pages'].isna())|
     (data['average_rating'].isna())|
     (data['published_year'].isna())]
```

```python
data_mising = data[~(data['description'].isna())&
     ~(data['num_pages'].isna())&
     ~(data['average_rating'].isna())&
     ~(data['published_year'].isna())]
```

```python
data_mising
```

```python
data_mising['categories'].value_counts().reset_index().sort_values(by='count', ascending=False)
```

```python
data_mising['words_in_description'] = data_mising['description'].apply(lambda x: len(str(x).split()))
data_mising['words_in_description']
```

```python
data_mising.loc[data_mising['words_in_description'].between(1, 4), 'description'] 
```

```python
data_mising_25_words =data_mising[data_mising['words_in_description'] > 25]
```

```python
data_mising_25_words['title_and_subtitle'] = np.where(
    data_mising_25_words['subtitle'].isna(),
    data_mising_25_words['title'],
    data_mising_25_words[['title', 'subtitle']].astype(str).agg(' : '.join, axis=1)
)
```

```python
data_mising_25_words
```

```python
data_mising_25_words['tagged_description'] = data_mising_25_words[['isbn13','description']].astype(str).agg(' '.join, axis=1)
data_mising_25_words
```

```python
(
    data_mising_25_words
    .drop(columns=['subtitle', 'missing-description', 'age-of-book','words_in_description'], axis=1)
    .to_csv('data_cleaned.csv', index=False)
)
```
